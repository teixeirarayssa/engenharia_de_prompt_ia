# **Calculadora Simples**

```python
def somar(num1, num2):
    # Função para realizar a operação de soma
    return num1 + num2

def subtrair(num1, num2):
    # Função para realizar a operação de subtração
    return num1 - num2

while True:
    # Loop principal da calculadora para permitir múltiplas operações
    print("\n--- Calculadora Simples ---")
    print("Escolha uma operação:")
    print("1. Soma (+)")
    print("2. Subtração (-)")
    print("3. Sair")

    # Solicita ao usuário que escolha uma operação
    escolha = input("Digite sua escolha (1/2/3): ")

    if escolha == '3':
        # Se o usuário escolher '3', a calculadora é encerrada
        print("Obrigado por usar a calculadora! Até mais.")
        break

    if escolha in ('1', '2'):
        # Verifica se a escolha é uma operação válida (soma ou subtração)
        try:
            # Solicita os números ao usuário e tenta convertê-los para float
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            # Captura erro se a entrada não for um número válido
            print("Entrada inválida. Por favor, digite números válidos.")
            continue # Continua para a próxima iteração do loop

        if escolha == '1':
            # Se a escolha for '1', realiza a soma
            resultado = somar(num1, num2)
            print(f"Você pediu para somar {num1} e {num2}. O resultado é {resultado}.")
        elif escolha == '2':
            # Se a escolha for '2', realiza a subtração
            resultado = subtrair(num1, num2)
            print(f"Você pediu para subtrair {num1} e {num2}. O resultado é {resultado}.")
    else:
        # Mensagem de erro para escolha inválida
        print("Opção inválida. Por favor, escolha 1, 2 ou 3.")
```
