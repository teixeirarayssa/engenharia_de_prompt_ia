```python
# Solicita ao usuário para inserir a temperatura e a converte para float.
# Este valor pode ser alterado para testar diferentes cenários (ex: 35.5 para superaquecimento).
temperatura = float(input("Por favor, insira a temperatura atual: "))

# Inicia uma estrutura condicional para verificar o valor da temperatura.
# Se a temperatura for maior que 30, o sistema está superaquecendo.
if temperatura > 30:
    # Se a condição for verdadeira, exibe uma mensagem de alerta de superaquecimento.
    # Separa o texto em aspas e as variáveis por vírgula.
    print("ALERTA: Temperatura do sistema (", temperatura, "°C) muito alta! Superaquecimento detectado.")
# Se a condição 'temperatura > 30' for falsa (ou seja, temperatura é 30 ou menor),
# executa o bloco de código abaixo.
else:
    # Exibe uma mensagem indicando que a temperatura está normal.
    # Separa o texto em aspas e as variáveis por vírgula.
    print("Temperatura do sistema: ", temperatura, "°C. Tudo normal.")
```
