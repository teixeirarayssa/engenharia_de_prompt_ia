```python
import time # Importa o módulo time para usar a função sleep para pausar a execução.
import random # Importa o módulo random para gerar números aleatórios (temperaturas).

def enviar_alerta(mensagem):
    """Simula o envio de um alerta para o console e por e-mail."""
    # Imprime a mensagem de alerta crítico no console, formatando-a.
    print(f"\nALERTA CRÍTICO: {mensagem}")
    # Simula o envio de um e-mail para administradores.
    print("E-mail simulado enviado para administradores: ALERTA CRÍTICO")
    # Em um cenário real, aqui você integraria com um serviço de e-mail (ex: smtplib)
    # ou um serviço de notificação (ex: Twilio para SMS, Slack, etc.).

def monitorar_sistema(limite_critico_temperatura=35.0, intervalo_verificacao=5):
    """
    Monitora continuamente a temperatura e dispara alertas se crítica.

    Args:
        limite_critico_temperatura (float): O limite de temperatura acima do qual um alerta é disparado.
        intervalo_verificacao (int): O tempo em segundos entre cada verificação de temperatura.
    """
    print("--- Iniciando monitoramento autônomo do sistema ---")
    # Informa o limite de temperatura que será monitorado.
    print(f"Monitorando temperatura com limite crítico: {limite_critico_temperatura}°C")
    # Informa o intervalo entre as verificações.
    print(f"Intervalo de verificação: {intervalo_verificacao} segundos.\n")

    try:
        # Loop infinito para monitorar continuamente o sistema.
        while True:
            # Simula a leitura de um sensor de temperatura.
            # Gera um valor de temperatura aleatório entre 20.0 e 40.0 graus Celsius.
            # Em um cenário real, você leria dados de um sensor hardware ou API.
            temperatura_atual = random.uniform(20.0, 40.0)

            # Exibe a temperatura atual formatada com uma casa decimal.
            print(f"Temperatura atual: {temperatura_atual:.1f}°C")

            # Verifica se a temperatura atual excedeu o limite crítico.
            if temperatura_atual > limite_critico_temperatura:
                # Constrói a mensagem de alerta com detalhes da temperatura e limite.
                mensagem_alerta = (
                    f"Temperatura do sistema ({temperatura_atual:.1f}°C) "
                    f"excedeu o limite crítico de {limite_critico_temperatura}°C!"
                )
                # Chama a função para enviar o alerta.
                enviar_alerta(mensagem_alerta)
            else:
                # Se a temperatura estiver normal, exibe uma mensagem indicando isso.
                print("Temperatura dentro dos limites normais.")

            # Pausa a execução pelo tempo definido em 'intervalo_verificacao' antes da próxima iteração.
            time.sleep(intervalo_verificacao)
    except KeyboardInterrupt:
        # Captura a interrupção do teclado (Ctrl+C) para encerrar o monitoramento de forma graciosa.
        print("\nMonitoramento interrompido pelo usuário.")
    except Exception as e:
        # Captura qualquer outra exceção inesperada que possa ocorrer.
        print(f"\nOcorreu um erro inesperado: {e}")

# Instruções para o usuário sobre como iniciar o monitoramento.
print("Para iniciar o programa autônomo, descomente a chamada à função 'monitorar_sistema()' e execute esta célula.")
print("Exemplo: 'monitorar_sistema(limite_critico_temperatura=33.0, intervalo_verificacao=10)'")
# monitorar_sistema(limite_critico_temperatura=33.0, intervalo_verificacao=10)
```
